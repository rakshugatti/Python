```python
import os

INPUT_FILE = "employee_logs.txt"
OUTPUT_FILE = "processed_logs.txt"

# ---------------------------------------------------
# Step 1: Create sample corporate log file
with open(INPUT_FILE, "w") as file:
    file.write("101,John,Login\n")
    file.write("102,Emma,Logout\n")
    file.write("103,Raj,Login\n")
    file.write("104,Sophia,Login\n")

# ---------------------------------------------------
# Step 2: Read entire file using read()
print("Full File Content using read():\n")
with open(INPUT_FILE, "r") as file:
    content = file.read()
    print(content)

# ---------------------------------------------------
# Step 3: Read first line using readline()
print("First record using readline():")
with open(INPUT_FILE, "r") as file:
    first_line = file.readline()
    print(first_line)

# ---------------------------------------------------
# Step 4: Read all lines using readlines()
print("All records using readlines():")
with open(INPUT_FILE, "r") as file:
    lines = file.readlines()
    print(lines)

# ---------------------------------------------------
# Step 5: Print line by line with line numbers
print("\nRecords with Line Numbers:")
with open(INPUT_FILE, "r") as file:
    for line_number, line in enumerate(file, start=1):
        print(f"{line_number}: {line.strip()}")

# ---------------------------------------------------
# Step 6: Write processed summary to new file
total_records = len(lines)

with open(OUTPUT_FILE, "w") as file:
    file.write("Employee Log Summary\n")
    file.write("---------------------\n")
    file.write(f"Total Records: {total_records}\n")

print("\nSummary written to processed_logs.txt successfully!")
```
