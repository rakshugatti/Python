```python
# Simple demonstration of Python memory behavior using id()

print("=== INTEGER CACHING ===")
a = 10
b = 10
print(id(a), id(b))   # Same

x = 300
y = 300
print(id(x), id(y))   # Often different 


print("\n=== STRING INTERNING ===")
s1 = "hello"
s2 = "hello"
print(id(s1), id(s2))   # Same → identical strings reused

s3 = "".join(["he", "llo"])  # created dynamically
print(id(s3), id(s1))        # May be different → not always reused


print("\n=== LIST (MUTABLE) ===")
lst1 = [1, 2, 3]
lst2 = [1, 2, 3]
print(id(lst1), id(lst2))    # Different → lists are separate objects

print("Before:", id(lst1))
lst1.append(4)
print("After :", id(lst1))   # Same → list is mutable
```

```python
1. Integers

Small numbers like 10 are stored only once → same ID.

Big numbers like 300 may be stored separately → different ID.

 2. Strings

"hello" is stored once → same ID.

Strings created during runtime (like "".join(...)) may not be reused → different ID.

 3. Lists

Lists are mutable, so:

Two lists with same values have different IDs.

Changing a list keeps the ID same.
```
