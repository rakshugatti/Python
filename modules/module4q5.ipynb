```python
#  dynamic typing using type() and id()

# Integer
x = 10
print("Value:", x, "Type:", type(x), "ID:", id(x))

# String
x = "Hello"
print("Value:", x, "Type:", type(x), "ID:", id(x))

# List
x = [1, 2, 3]
print("Value:", x, "Type:", type(x), "ID:", id(x))

# Dictionary
x = {"name": "Rakshitha", "age": 21}
print("Value:", x, "Type:", type(x), "ID:", id(x))
```
