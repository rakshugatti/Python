```python
# Integer
a = 10
print(a, type(a))
print("Before:", a, "ID:", id(a))
a = a + 5
print("After :", a, "ID:", id(a))   
```

```python
# Float
b = 10.5
print(b, type(b))
print("Before:", b, "ID:", id(b))
b = b * 2
print("After :", b, "ID:", id(b)) 
```

```python
# Boolean
e = True
print(e, type(e))
```

```python
# Complex
c = 3 + 4j
print(c, type(c))
```

```python
# String
d = "Hello, Python"
print(d, type(d))

```

```python
# List
f = [1, 2, 3, 4]
print(f, type(f))
```

```python
# Tuple
g = (10, 20, 30)
print(g, type(g))
print("Before ID:", id(g))
g = g + (4,)
print("After ID :", id(g))   
```

```python
# Set
h = {1, 2, 3, 4}
print(h, type(h))
print("Before:", h, "ID:", id(h))
h.add(4)
print("After :", h, "ID:", id(h)) 
```

```python
# Dictionary
i = {"name": "Rakshitha", "age": 21}
print(i, type(i))
print("Before:", i, "ID:", id(i))
i["age"] = 20
print("After :", i, "ID:", id(i)) 
```

```python
# Frozenset
j = frozenset([1, 2, 3, 4])
print(j, type(j))
```

```python
# Bytes
k = bytes([10, 20, 30, 40])
print(k, type(k))
```

```python
# Bytearray
l = bytearray([5, 10, 15])
print(l, type(l))
```

```python
 #NoneType
m = None
print(m, type(m))
```
