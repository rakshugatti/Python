```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    "Year": [2019, 2020, 2021, 2022, 2023],
    "Sales_Units": [5000, 7000, 12000, 20000, 35000],
    "Revenue_Million": [50, 75, 140, 250, 500],
    "Profit_Margin": [5, 6, 8, 10, 15]
}

data
```

```python
ef_df=pd.DataFrame(data)
```

```python
ef_df
```

```python
plt.scatter(ef_df['Sales_Units'],ef_df['Revenue_Million'])
```

```python
plt.scatter(ef_df['Sales_Units'],ef_df['Revenue_Million'])
plt.xlabel('Sales_Units')
plt.ylabel('Revenue_Million')
plt.title('EV Sales Trend')
plt.show()
```

```python
plt.plot(ef_df['Year'], ef_df['Sales_Units'])
plt.xlabel('Year')
plt.ylabel('Sales Units')
plt.title('EV Sales Trend')
plt.show()
```

```python
plt.plot(ef_df['Year'], ef_df['Sales_Units'])
```

```python
colors = ['red', 'green', 'blue', 'orange', 'purple']

plt.plot(
    ef_df['Year'], 
    ef_df['Sales_Units'], 
    marker='o'
)

for i in range(len(ef_df['Year']) - 1):
    plt.plot(
        ef_df['Year'].iloc[i:i+2],
        ef_df['Sales_Units'].iloc[i:i+2],
        color=colors[i % len(colors)]
    )

plt.xlabel('Year')
plt.ylabel('Sales Units')
plt.title('EV Sales Trend')
plt.show()
```

```python
plt.bar(ef_df["Year"], ef_df["Revenue_Million"])
plt.xlabel("Year")
plt.ylabel("Revenue (Million)")
plt.title("Year-wise Revenue Comparison")
plt.show()
```

```python
plt.boxplot(ef_df["Profit_Margin"])
plt.title("EV Sales Growth Over Year")
plt.ylabel('Profit_Margin(%)')
plt.show()
```

```python
categories = ['Sales', 'Revenue', 'Profit', 'Stations']
values = [35000, 500, 15, 3000]

# Correct spelling: linspace instead of linespace
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)

# Close the loop
values = np.concatenate((values, [values[0]]))
angles = np.concatenate((angles, [angles[0]]))

# Create figure correctly
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# Plot
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)

# Correct tick labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title("EV Company Performance")
plt.show()#rader chart
```
